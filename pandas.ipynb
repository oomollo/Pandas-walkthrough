```python
#join -types of joins
#inner join-returns only records with matching keys in both tables
#outer join-returns all records,when theres a match in atleast 1 table, unmatched tables will return NaN
#left join-returns all records rom the left table and matching rows from the right table, unmstched rows will have NaN
#right join-returns all records rom the right table and matching rows from the left table, unmstched rows will have NaN
```

```python
df1= pd.DataFrame({'key':['A','B','C'], 'value1':[1,2,3]})
df2= pd.DataFrame({'key':['B','C','D'], 'value2':[4,5,6]})
#innerjoin
inner_join= pd.merge(df1,df2, on='key',how='inner')
inner_join
```

```python
#outer join
outer_join= pd.merge(df1,df2, on='key',how='outer')
print(outer_join)
```

```python
#left join
left_join= pd.merge(df1,df2, on='key',how='left')
print('Left Join:\n',left_join)

#right join
right_join= pd.merge(df1,df2, on='key',how='right')
print('Right Join:\n',right_join)
```
